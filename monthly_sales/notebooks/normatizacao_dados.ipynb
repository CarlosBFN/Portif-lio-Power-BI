```python
'''
Transformações a serem feitas
  Units Sold            obj > int
  Manufacturing Price   int > float
  Sale Price            int > float
  Gross Sales           obj > int
  Discounts             obj > float
  Sales                 obj > float
  COGS                  obj > float
  Profit                obj > float
  Date                  obj > date
'''

# Apagar espaços nos nomes das colunas
base_13_14.columns = base_13_14.columns.str.replace(' ', '')

# Função de susbtituição de vírgulas por pontos nas strings
def preparar_float(virgula):
  virgula = virgula.str.replace(',', '.')
  return virgula

# Aplicação da função de substituição de vírgulas e conversão das strings em floats
base_13_14 [["Discounts","Sales","COGS","Profit"]] = base_13_14 [["Discounts","Sales","COGS","Profit"]].apply(preparar_float)
base_13_14 [["Discounts","Sales","COGS","Profit"]] = base_13_14 [["Discounts","Sales","COGS","Profit"]].astype(float)

#Conversão das ints em floats
base_13_14 [["ManufacturingPrice","SalePrice"]] = base_13_14 [["ManufacturingPrice","SalePrice"]].astype(float)

#Conversão do restante das colunas para float
base_13_14 [["UnitsSold","GrossSales"]] = base_13_14 [["UnitsSold","GrossSales"]].apply(preparar_float)
base_13_14 [["UnitsSold","GrossSales"]] = base_13_14 [["UnitsSold","GrossSales"]].astype(float)


```

```python
# Correção da data
base_13_14 ["Date"] = base_13_14 ["Date"].str.replace('.', '')

def corrigir_datas(datas):
    datas = datas.replace('jan', 'Jan').replace('fev','Feb').replace('mar', 'Mar').replace('abr', 'Apr').replace('mai', 'Mai').replace('jun', 'Jun').replace('jul', 'Jul').replace('ago', 'Aug').replace('set', 'Sep').replace('out', 'Oct').replace('nov','Nov').replace('dez','Dec')
    return datas

base_13_14 ["Date"] = base_13_14 ["Date"].apply(corrigir_datas)

date_string = "1-set-16"
print("date_string =", date_string)

date_object = datetime.strptime(date_string, "%d-%b-%Y")

print("date_object =", date_object)
```
