```python
import pandas as pd

metas_vendas = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSR9B8Y-1zeA2SRxlrzstVyTanY3s7rVwmBONk4nGCCCGM3NAxgvcd4jNtIxXekig/pub?output=csv")
resultados = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRsU5ZPhhUVRtXGMXzz6ykJ6wnCd9VqZIe0xA8VU9VxFTixrchP1fmDWvkhgzhoJcWnqi9Y_lY3JyL6/pub?output=csv")

venda_descreditada = pd.DataFrame({
    'Unnamed: 0': [''],
    'Nome': ['Venda descreditada'],
    'Email': [''],
    'Canal': [''],
    'Dias a trabalhar': [''],
    'Meta diária': [''],
    'Meta mensal': ['']})
```

```python
metas_vendas
```

```python
metas_vendas = metas_vendas.drop(["Unnamed: 0"], axis=1)

metas_vendas.head()
```

```python
metas_vendas.info()
```

```python
metas_vendas['Meta diária'] = metas_vendas['Meta diária'].str.replace(',','.').astype(float)
metas_vendas['Meta mensal'] = metas_vendas['Meta mensal'].str.replace(',','.').astype(float)
metas_vendas.info()
```

```python
resultados.isna().sum()
```

```python
##resultados = resultados.dropna()
resultados ["Nome"].fillna("Venda descreditada", inplace=True)

resultados
```

```python
resultados.info()
```

```python
resultados["Nome"] = resultados["Nome"].str.replace('ê','e')
##resultados["Nome"] = resultados["Nome"].str.replace('Luiz Gustavo Araujo','Luiz Gustavo Souza')

resultados
```

```python
resultados['date']= pd.to_datetime(resultados['date'])

resultados.info()
```

```python
resultados_intermediário = resultados.groupby(['Nome']).count()

resultados_intermediário.head()
```

```python
##resultados_soma = pd.merge(metas_vendas, resultados_intermediário, how="right", on="Nome")

resultados_soma = pd.merge(metas_vendas, resultados_intermediário, how="left", on="Nome")
resultados_soma = resultados_soma.rename({'date': 'Resultado entregue'}, axis = 1)

resultados_soma
```

```python
##resultados_soma ["Meta mensal novembro"] = resultados_soma ["Resultado entregue"] * 1.1
##resultados_soma ["Meta diária novembro"] = resultados_soma ["Meta diária"] * 1.1
##resultados_soma ["Restante novembro"] = resultados_soma ["Meta mensal novembro"] - resultados_soma ["Resultado entregue"]

##resultados_soma.to_excel('Etapa 4.xlsx')
```

```python
resultados_soma ["Entrega/dia"] = resultados_soma ["Resultado entregue"] / resultados_soma ["Dias a trabalhar"]
pd.options.display.float_format = '{:.1f}'.format

resultados_soma
```

```python
resultados_soma ["Meta batida"] = resultados_soma ["Resultado entregue"] >= resultados_soma ["Meta mensal"]

resultados_soma
```

```python
resultados_soma ["75% da meta"] = resultados_soma ["Resultado entregue"] >= resultados_soma ["Meta mensal"] * 0.75
##resultados_soma ["125% da meta"] = resultados_soma ["Resultado entregue"] >= resultados_soma ["Meta mensal"] * 1.25


resultados_soma
```
