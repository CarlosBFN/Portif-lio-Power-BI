Análise e normatização da tabela "vendedores"



*   Verificação dos "Dtype"
*   Verificação de outras funções diferentes de vendedores
*   Contagem de dias a se trabalhar


```python
vendedores
```

```python
def corrigir_nomes(nome):
    nome = nome.replace('ç', 'c').replace('á','a').replace('â', 'a').replace('ã', 'a').replace('ô', 'o').replace('õ', 'o').replace('ó', 'o').replace('é', 'e').replace('ê', 'e').replace('ú', 'u').replace('í','i')
    return nome

vendedores ['Nome'] = vendedores ['Nome'].apply(corrigir_nomes)
vendedores ['Email'] = vendedores ['Email'].apply(corrigir_nomes)

vendedores
```

```python
vendedores.info()
```

```python
vendedores["Funcao"].value_counts()
```

```python
vendedores_pura = (vendedores["Funcao"] == "Vendedor")
vendedores_normatizada = vendedores [vendedores_pura]
```

```python
vendedores_normatizada["Funcao"].value_counts()
```

```python
vendedores_normatizada ["Dias a trabalhar"] = (vendedores_normatizada ["folga_1"] + vendedores_normatizada ["folga_2"] - 30) * -1

vendedores_normatizada.info()
```

**Merge das tabelas, normatização e amostra para verificação**





```python
metas_vendas = pd.merge(vendedores_normatizada, metas, on="Canal")
metas_vendas = metas_vendas.rename({'Produtividade': 'Meta diária'}, axis = 1)

metas_vendas.head()
```

```python
metas_vendas ["Meta mensal"] = metas_vendas ["Meta diária"].astype(float) * metas_vendas ["Dias a trabalhar"].astype(float)
metas_vendas.sample(20)
```

```python
metas_vendas = metas_vendas [['Nome','Email','Canal','Dias a trabalhar','Meta diária','Meta mensal']]

metas_vendas
```
